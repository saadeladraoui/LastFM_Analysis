### Extracting songs features using the Spotify API

We use the Spotipy library to connect to the Spotipy API. 

All information about getting a client ID and a secret key can be found by crating an account on [Spotify Developers](https://developer.spotify.com/dashboard/login).

```python
import spotipy
import spotipy.util as util

#Put CLIENT ID
CLIENT_ID = '###########################' 
#Put CLIENT SECRET
CLIENT_SECRET = '###########################'


token = util.oauth2.SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
cache_token = token.get_access_token()
sp = spotipy.Spotify(cache_token)
```

We drop the duplicates songs from our data.

```python
df_songs = df[['artist','name','song']].drop_duplicates()
df_songs['uri']=None
df_songs['album']=None
```

The function search_song searches a song by artist and track name then returns the spotify uri of the song and the album of the song.

We scrap album names one more time because a consistent number my old scrobbles don't have a correct album name.

```python
def search_song(artist_name,track):
    result = sp.search(q='artist:' + artist_name.replace("'","")+' track:'+track.replace("'",""), type='track')
    return (result['tracks']['items'][0]['uri'],result['tracks']['items'][0]['album']['name'])
```

```python
songs_not_found = []
for song in df_songs.values:
    try:
        result = search_song(song[0],song[1])
        df_songs.loc[df_songs.song==song[2],'uri']=result[0]
        df_songs.loc[df_songs.song==song[2],'album']=result[1]
    except:
        songs_not_found.append(song[2])
```

```python
#now that we scrapped URIs we store the dataframe using pickle
df_songs.to_pickle('data/df_songs.pkl')
```

```python
#df_songs=pd.read_pickle('df_songs.pkl')
df_songs=df_songs[df_songs.uri.notnull()]
print "The number unique songs with an URI is : " +str(df_songs.count()[0]) 
```

We create columns for the features we're going to extract.

```python
df_songs['duration_ms']=None
df_songs['energy']=None
df_songs['speechiness']=None
df_songs['loudness']=None
df_songs['danceability']=None
df_songs['acousticness']=None
df_songs['valence']=None
df_songs['tempo']=None
df_songs['instrumentalness']=None
df_songs['liveness']=None
df_songs['mode']=None
df_songs['time_signature']=None
```

We loope through the songs and extract the Spotify features from each one.

```python
i=1
for song in df_songs.values:
    features = sp.audio_features(str(song[3]))
    try:
        features_array=[features[0]['acousticness'],features[0]['danceability'],features[0]['duration_ms'],features[0]['energy'],\
                        features[0]['instrumentalness'],features[0]['liveness'],features[0]['loudness'],features[0]['speechiness'],\
                        features[0]['tempo'],features[0]['valence'],features[0]['mode'],features[0]['time_signature']]
        df_songs.loc[df_songs.uri==song[3],['acousticness','danceability','duration_ms','energy','instrumentalness',\
                                            'liveness','loudness','speechiness','tempo','valence','mode',\
                                           'time_signature']]=features_array
    except Exception as e:
        print e
        break
    
    if i%300==0:
        time.sleep(1)
    i=i+1
```

```python
#We save the resulting dataframe 
df_songs.to_pickle('data/df_songs_features.pkl')
```

### Evolution of features in time 

We merge songs features with all my scrobbles of the last 6 years.

```python
df=pd.read_pickle('data/all_data.pkl')
df['song']=df['artist'] + ' - ' + df['name'] 
df['date']=pd.to_datetime(df['date'], format='%d %b %Y, %H:%M')
```

```python
df_all=pd.merge(df, df_songs, on='song', how='left')
df_all=df_all[df_all.loudness.notnull()]
```

```python
df_all.head(3)
```

We group data by month and calculate the mean of each month for every feature.

```python
df_plot_features=df_all.set_index('date').groupby([pd.TimeGrouper('M')])\
['valence','energy','loudness','danceability','tempo','instrumentalness','acousticness'].agg('mean')
```

We normalize the Tempo value.

```python
x = df_plot_features[['tempo']].values.astype(float)
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
```

We normalize all values between 0 and 1 using a Min-Max Normalization.

```python
normalized_df=(df_plot_features-df_plot_features.min())/(df_plot_features.max()-df_plot_features.min())
```

Let's plot the evolution of features during the months.

```python
normalized_df.plot(figsize=(20, 12))
plt.show()
```

This line graph points out my musical taste has changed starting from 2017.

- I started listened to less acoustic songs, the decline is very visible.


- I started listened to more high energy and fast tempo songs which correlates on what we sang on the analysis of types music in Part 2, we found out that I started listening to more and more Punk and Post-Punk starting from 2017.


- In terms of valence, it has been consistent over the years. Maybe, we can say that my songs a bit more happy in 2018.

When plotting using a cumsum, we can how acousticness stopped growing while energy and tempo had a big increase.

```python
normalized_df.cumsum().plot(figsize=(14, 8))
plt.show()
```
